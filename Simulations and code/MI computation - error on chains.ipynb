# Error mutual information from convolution

## 2 chain

```python
def MIError(x,y):
    return -0.5 * np.log(1 - x**2 * y**2) - 0.25 * np.log(1 - x**2) * np.log(1 - y**2)

xx, yy = np.meshgrid(np.linspace(0,1,3_000), np.linspace(0,1,3_000))

# fig,ax = plt.subplots(subplot_kw={"projection": "3d"})
zz = MIError(xx,yy)
zz[np.abs(zz) > .1] = np.nan

fig,ax = plt.subplots(figsize=(4,3))
ax.set_xlabel(r"$\rho_{1,2}$")
ax.set_ylabel(r"$\rho_{2,3}$")
r = ax.contourf(xx,yy,zz, 15)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(r, cax=cax)
fig.tight_layout()

# plt.savefig(figures_path + "2 chain error - MI.pdf", bbox_inches='tight')
# plt.savefig(figures_path + "2 chain error - MI", bbox_inches='tight')
```

## 3 chain

```python
def MIError(x,y, rho3 = 0.7):
    return -0.5 * np.log(1 - x**2 * y**2 * rho3**2) - (-0.5)**3 * np.log(1 - x**2) * np.log(1 - y**2) * np.log(1 - rho3**2)

xx, yy = np.meshgrid(np.linspace(0,1,3_000), np.linspace(0,1,3_000))
zz = MIError(xx,yy)
zz[np.abs(zz) > .1] = np.nan



fig,ax = plt.subplots(figsize=(4,3))
ax.set_xlabel(r"$\rho_{1,2}$")
ax.set_ylabel(r"$\rho_{2,3}$")
r = ax.contourf(xx,yy,zz, 15)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(r, cax=cax)
fig.tight_layout()

# plt.savefig(figures_path + "3 chain error - MI - rho3 0_7.pdf", bbox_inches='tight')
# plt.savefig(figures_path + "3 chain error - MI - rho3 0_7", bbox_inches='tight')
```

## 4 chain

```python
def MIError(x,y, rho3 = 0.85, rho4 = 0.6):
    return -0.5 * np.log(1 - x**2 * y**2 * rho3**2 * rho4**2) - (-0.5)**4 * np.log(1 - x**2) * np.log(1 - y**2) * np.log(1 - rho3**2) * np.log(1 - rho4**2)

xx, yy = np.meshgrid(np.linspace(0,1,3_000), np.linspace(0,1,3_000))
zz = MIError(xx,yy)
zz[np.abs(zz) > .1] = np.nan


fig,ax = plt.subplots(figsize=(4,3))
ax.set_xlabel(r"$\rho_{1,2}$")
ax.set_ylabel(r"$\rho_{2,3}$")
r = ax.contourf(xx,yy,zz, 15)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(r, cax=cax)
fig.tight_layout()

# plt.savefig(figures_path + "4 chain error - MI - rho3 0_85 rho4 0_6.pdf", bbox_inches='tight')
# plt.savefig(figures_path + "4 chain error - MI - rho3 0_85 rho4 0_6", bbox_inches='tight')
```
