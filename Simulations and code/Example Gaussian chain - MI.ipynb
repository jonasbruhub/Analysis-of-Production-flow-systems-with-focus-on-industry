# Gaussian chain example - Triangular - MI

```python
N = 10

# cor = np.eye(N, k = 1) * 0.9

# the first element of the vector multiplied on is non-important as is multiplied with the first column of 0
cor = np.eye(N, k = 1) * np.array([0,
                                   0.6,
                                   0.5,
                                   0.4,
                                   0.2,
                                   0.9,
                                   0.8,
                                   0.9,
                                   0.8,
                                   0.7])



var_names = GenLabels("X", N)
G_obs = cor @ np.linalg.pinv(np.eye(N) - cor)

G_obs = -0.5 * np.log(1 - G_obs**2)

# np.fill_diagonal(G_obs, 0)



G_dir = G_obs @ np.linalg.pinv( np.eye(N) + G_obs)
fig, ax = plt.subplots(1,1, figsize = (4,3))
r = ax.matshow(G_dir, cmap = "Reds", vmin = 0)
ax.set_xticks(range(N))
ax.set_xticklabels(var_names)
ax.set_yticks(range(N))
ax.set_yticklabels(var_names)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(r, cax=cax);
fig.tight_layout()
# plt.show()

plt.savefig(figures_path + "G dir from triangular G obs - MI.pdf", bbox_inches='tight')
plt.savefig(figures_path + "G dir from triangular G obs - MI", bbox_inches='tight')



fig, ax = plt.subplots(1,1, figsize = (4,3))
r = ax.matshow(G_obs, cmap = "Reds", vmin = 0)
ax.set_xticks(range(N))
ax.set_xticklabels(var_names)
ax.set_yticks(range(N))
ax.set_yticklabels(var_names)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(r, cax=cax);
fig.tight_layout()
# plt.show()

plt.savefig(figures_path + "triangular G obs - MI.pdf", bbox_inches='tight')
plt.savefig(figures_path + "triangular G obs - MI", bbox_inches='tight')


fig, ax = plt.subplots(1,1, figsize = (4,3))
gr = nx.from_numpy_array(G_dir > 2.1e-2, create_using = nx.DiGraph)
# cut-offs: 2e-2, 2.1e-2, 4.51e-2
    # correlation between 4-5 is 0.2


# nx.draw(gr, pos=graphviz_layout(gr), node_size = 500, with_labels= True, ax = ax, node_color='lightgreen')
pos = graphviz_layout(gr)
pos = nx.circular_layout(gr)

nx.draw(gr, pos = pos, node_size = 500, with_labels= True, ax = ax, node_color='#93F592', labels = { k:v for (k,v) in zip(range(N), var_names)} )
fig.tight_layout()
# plt.show()

#  - cutoff 4_51e-2
#  - cutoff 2e-2
#  - cutoff 2_1e-2
# plt.savefig(figures_path + "Chain graph from triangular G obs - MI - cutoff 2_1e-2.pdf", bbox_inches='tight')
# plt.savefig(figures_path + "Chain graph from triangular G obs - MI - cutoff 2_1e-2", bbox_inches='tight')
```

# Gaussian chain example - Symmetric - MI

```python
N = 10

# cor = np.eye(N, k = 1) * 0.9

# the first element of the vector multiplied on is non-important as is multiplied with the first column of 0
cor = np.eye(N, k = 1) * np.array([0,
                                   0.6,
                                   0.5,
                                   0.4,
                                   0.2,
                                   0.9,
                                   0.8,
                                   0.9,
                                   0.8,
                                   0.7])



var_names = GenLabels("X", N)
G_obs = cor @ np.linalg.pinv(np.eye(N) - cor)

G_obs = G_obs + G_obs.T

G_obs *= 1/0.9226338713927753 * 0.495

# np.fill_diagonal(G_obs, 0)



G_dir = G_obs @ np.linalg.pinv( np.eye(N) + G_obs)
fig, ax = plt.subplots(1,1, figsize = (5,4))
r = ax.matshow(G_dir, cmap = "Reds", vmin = 0)
ax.set_xticks(range(N))
ax.set_xticklabels(var_names)
ax.set_yticks(range(N))
ax.set_yticklabels(var_names)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(r, cax=cax);
fig.tight_layout()
# plt.show()

plt.savefig(figures_path + "G dir from symmetric G obs.pdf", bbox_inches='tight')
plt.savefig(figures_path + "G dir from symmetric G obs", bbox_inches='tight')



fig, ax = plt.subplots(1,1, figsize = (5,4))
r = ax.matshow(G_obs, cmap = "Reds", vmin = 0)
ax.set_xticks(range(N))
ax.set_xticklabels(var_names)
ax.set_yticks(range(N))
ax.set_yticklabels(var_names)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(r, cax=cax);
fig.tight_layout()
# plt.show()

plt.savefig(figures_path + "symmetric G obs.pdf", bbox_inches='tight')
plt.savefig(figures_path + "symmetric G obs", bbox_inches='tight')


fig, ax = plt.subplots(1,1, figsize = (5,4))
gr = nx.from_numpy_array(G_dir > 2.e-1, create_using = nx.DiGraph)

# nx.draw(gr, pos=graphviz_layout(gr), node_size = 500, with_labels= True, ax = ax, node_color='lightgreen')
pos = graphviz_layout(gr)
pos = nx.circular_layout(gr)

nx.draw(gr, pos = pos, node_size = 500, with_labels= True, ax = ax, node_color='#93F592', labels = { k:v for (k,v) in zip(range(N), var_names)} )
fig.tight_layout()
# plt.show()

plt.savefig(figures_path + "Chain graph from symmetric G obs.pdf", bbox_inches='tight')
plt.savefig(figures_path + "Chain graph from symmetric G obs", bbox_inches='tight')
```
